**factory Scenario

```python
production = np.array([
    [550, 590, 480, 570],   # Gadgets
    [400, 420, 450, 470],   # Tools
    [480, 510, 530, 550]    # Machines
])
```

```python
production = production * 1.05
print(production)
```

```python
gadgets = production[0]
print("Gadgets Production:", gadgets)
```

```python
avg_per_product = np.mean(production, axis=1)
print("Average per product:", avg_per_product)
```

```python
high_values = production[production > 500]
print("Production values > 500:", high_values)
```
