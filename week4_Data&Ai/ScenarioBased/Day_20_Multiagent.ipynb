```python
# 🎯 Example 1: Sequential Multi-Agent System (Pipeline)
# 🧠 Scenario

# “A travel company has different employees (agents):

# Planner → decides steps

# Flight Agent → finds flights

# Weather Agent → checks weather

# Decision Agent → gives final answer”


def planner_agent(user_query):
    print("\n[Planner Agent] Creating plan...")
    return ["flight", "weather", "decision"]


def flight_agent():
    print("\n[Flight Agent] Fetching flights...")
    return [
        {"airline": "IndiGo", "price": 4500},
        {"airline": "Air India", "price": 5200}
    ]


def weather_agent():
    print("\n[Weather Agent] Checking weather...")
    return {"condition": "Clear", "temp": 28}


def decision_agent(flights, weather):
    print("\n[Decision Agent] Making decision...")

    cheapest = min(flights, key=lambda x: x["price"])

    if weather["condition"] == "Rain":
        return "Avoid travel due to bad weather"

    return f"Book {cheapest['airline']} at ₹{cheapest['price']}"


def travel_multi_agent(user_query):
    print("User Query:", user_query)

    plan = planner_agent(user_query)

    flights = None
    weather = None

    for step in plan:
        if step == "flight":
            flights = flight_agent()

        elif step == "weather":
            weather = weather_agent()

        elif step == "decision":
            result = decision_agent(flights, weather)

    return result


response = travel_multi_agent("Plan my trip Delhi to Mumbai")
print("\nFinal Answer:", response)
```

Scenario
“A hospital uses different employees (agents) to handle patient care in sequence.”

================================
AGENT 1: Intake Agent (Planner)
- Collects patient symptoms and history
- Decides which steps are needed (tests, consultations, etc.)

================================
AGENT 2: Diagnostic Agent
- Orders lab tests or scans
- Interprets results and identifies possible conditions

================================
AGENT 3: Treatment Agent
- Suggests treatment options (medication, therapy, surgery)
- Considers patient preferences and medical guidelines

================================
AGENT 4: Decision Agent
- Reviews all inputs (history, diagnostics, treatment options)
- Provides the final recommendation to the patient

```python

def intake_agent(patient_query):
    print("\n[Intake Agent] Collecting patient details...")

    # Simulated extraction (in real case → NLP/LLM)
    data = {
        "symptoms": ["fever", "cough"],
        "history": ["diabetes"],
        "next_steps": ["blood_test", "xray"]
    }

    return data

def diagnostic_agent(intake_data):
    print("\n[Diagnostic Agent] Running diagnostics...")

    # Simulated test results
    tests = {
        "blood_test": "normal",
        "xray": "infection detected"
    }

    # Simple rule-based diagnosis
    if "infection detected" in tests["xray"]:
        condition = "pneumonia"
    else:
        condition = "unknown"

    return {
        "tests": tests,
        "condition": condition
    }


def treatment_agent(diagnosis):
    print("\n[Treatment Agent] Suggesting treatments...")

    if diagnosis["condition"] == "pneumonia":
        treatments = ["antibiotics", "rest", "hydration"]
    else:
        treatments = ["further testing"]

    return {
        "treatment_options": treatments
    }

def decision_agent(intake, diagnosis, treatment):
    print("\n[Decision Agent] Finalizing recommendation...")

    condition = diagnosis["condition"]
    options = treatment["treatment_options"]

    if condition == "pneumonia":
        return f"Diagnosis: {condition}. Start {options[0]} for 7 days with rest."
    else:
        return "Condition unclear. Recommend further tests."


def hospital_multi_agent(patient_query):
    print("Patient Query:", patient_query)

    # Step 1: Intake
    intake_data = intake_agent(patient_query)

    # Step 2: Diagnosis
    diagnosis = diagnostic_agent(intake_data)

    # Step 3: Treatment
    treatment = treatment_agent(diagnosis)

    # Step 4: Decision
    final_decision = decision_agent(intake_data, diagnosis, treatment)

    return final_decision


response = hospital_multi_agent("I have fever and cough for 3 days")
print("\nFinal Recommendation:", response)
```

```python
# Example 2: Manager–Worker Multi-Agent System
# 🧠 Scenario

# “Now instead of fixed flow, we introduce a Manager Agent
# that assigns tasks dynamically to worker


def flight_agent():
    print("[Flight Agent] Working...")
    return [{"airline": "IndiGo", "price": 4500},
            {"airline": "Air India", "price": 5200}]


def weather_agent():
    print("[Weather Agent] Working...")
    return {"condition": "Clear", "temp": 28}


def manager_agent(user_query):
    print("\n[Manager Agent] Analyzing task...")

    tasks = []

    if "trip" in user_query.lower():
        tasks = ["flight", "weather"]

    return tasks


def run_system(user_query):
    print("User Query:", user_query)

    tasks = manager_agent(user_query)

    results = {}

    for task in tasks:
        if task == "flight":
            results["flights"] = flight_agent()

        elif task == "weather":
            results["weather"] = weather_agent()

    # Final decision
    cheapest = min(results["flights"], key=lambda x: x["price"])

    return f"Manager Decision: Book {cheapest['airline']} at ₹{cheapest['price']}"


# RUN
response = run_system("Plan my trip")
print("\nFinal Answer:", response)





```

Scenario: Corporate Market Research & Strategy
A company wants to explore launching a new product in a competitive market. The Manager Agent oversees the process and delegates tasks to specialized workers.

👩‍💼 Manager Agent
- Receives the overall goal: “Evaluate feasibility of launching Product Y in Asia.”
- Dynamically assigns tasks to worker agents depending on what’s needed.
- Example: If budget is unclear → send to Finance Worker. If regulations are complex → send to Legal Worker.

================================
📊 Worker Agents
================================
- Market Research Worker
- Collects competitor data, customer preferences, and demand forecasts.
- Reports: “High demand in Tier‑1 cities, moderate competition.”
- Finance Worker
- Analyzes budget, ROI, and pricing strategy.
- Reports: “Estimated $10M investment, ROI in 2 years.”
- Operations Worker
- Evaluates supply chain, production capacity, and logistics.
- Reports: “Factories can scale up, but shipping costs are high.”
- Legal Worker
- Reviews compliance, intellectual property, and regional regulations.
- Reports: “Trademark available, but import laws require certification.”
- HR Worker
- Assesses staffing needs and training requirements.
- Reports: “Need 50 new hires for customer support and sales.”


```python


# CORPORATE MARKET RESEARCH & STRATEGY

from openai import OpenAI
import json
import re

client = OpenAI(
    api_key="api key",  
    base_url="https://api.groq.com/openai/v1"
)


def safe_json_parse(text):
    try:
        return json.loads(text)
    except:
        match = re.search(r"\{.*\}|\[.*\]", text, re.DOTALL)
        if match:
            try:
                return json.loads(match.group())
            except:
                pass
    return {"error": "Invalid JSON", "raw": text}



def call_llm(system_prompt, user_message):
    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",  
        temperature=0,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user",   "content": user_message}
        ]
    )
    return response.choices[0].message.content.strip()


def market_research_worker(goal):
    print("\n  [Market Research Worker] Collecting data...")
    reply = call_llm(
        system_prompt="""
        Return ONLY valid JSON:
        {
          "demand_level": "High/Medium/Low",
          "target_cities": ["city1","city2","city3"],
          "competition_level": "High/Moderate/Low",
          "summary": "string"
        }
        """,
        user_message=goal
    )
    return safe_json_parse(reply)


def finance_worker(goal):
    print("\n  [Finance Worker] Analyzing budget & ROI...")
    reply = call_llm(
        system_prompt="""
        Return ONLY valid JSON:
        {
          "estimated_investment_usd": 1000000,
          "projected_roi_years": 3,
          "risk_level": "Low/Medium/High",
          "summary": "string"
        }
        """,
        user_message=goal
    )
    return safe_json_parse(reply)


def operations_worker(goal):
    print("\n  [Operations Worker] Evaluating supply chain...")
    reply = call_llm(
        system_prompt="""
        Return ONLY valid JSON:
        {
          "factory_scalability": "High/Medium/Low",
          "shipping_bottleneck": true,
          "bottleneck_reason": "string",
          "summary": "string"
        }
        """,
        user_message=goal
    )
    return safe_json_parse(reply)


def legal_worker(goal):
    print("\n  [Legal Worker] Reviewing regulations...")
    reply = call_llm(
        system_prompt="""
        Return ONLY valid JSON:
        {
          "trademark_available": true,
          "import_certification_required": false,
          "certification_name": "string",
          "summary": "string"
        }
        """,
        user_message=goal
    )
    return safe_json_parse(reply)


def hr_worker(goal):
    print("\n  [HR Worker] Assessing staffing needs...")
    reply = call_llm(
        system_prompt="""
        Return ONLY valid JSON:
        {
          "new_hires_required": 50,
          "roles": {"engineer": 20, "sales": 30},
          "summary": "string"
        }
        """,
        user_message=goal
    )
    return safe_json_parse(reply)



def manager_agent(goal):
    print("\n[Manager Agent] Analyzing goal...")
    print(f"  Goal: {goal}\n")

    reply = call_llm(
        system_prompt="""
        Return ONLY JSON list:
        ["market_research", "finance", "operations", "legal", "hr"]
        """,
        user_message=goal
    )

    tasks = safe_json_parse(reply)
    print(f"  Tasks Assigned: {[t.replace('_', ' ').title() for t in tasks]}")
    return tasks


def make_final_decision(results):
    score, flags, positives = 0, [], []

    mr  = results.get("market_research", {})
    fin = results.get("finance", {})
    ops = results.get("operations", {})
    leg = results.get("legal", {})
    hr  = results.get("hr", {})

    if mr.get("demand_level") == "High":
        score += 1; positives.append("Strong demand")

    if fin.get("projected_roi_years", 99) <= 3:
        score += 1; positives.append("Good ROI")

    if ops.get("factory_scalability") == "High":
        score += 1; positives.append("Scalable operations")

    if leg.get("trademark_available"):
        score += 1; positives.append("Trademark available")

    if hr.get("new_hires_required", 999) <= 100:
        score += 1; positives.append("Manageable hiring")

    if ops.get("shipping_bottleneck"):
        flags.append("Shipping bottleneck")

    if leg.get("import_certification_required"):
        flags.append("Certification needed")

    if fin.get("risk_level") == "High":
        flags.append("High financial risk")

    pct = (score / 5) * 100
    verdict = "PROCEED" if pct >= 75 else "CONDITIONAL" if pct >= 50 else "NOT RECOMMENDED"

    return {"score": f"{pct:.0f}%", "verdict": verdict, "positives": positives, "flags": flags}


def run_system(goal):
    print("  CORPORATE MARKET RESEARCH & STRATEGY SYSTEM")

    tasks = manager_agent(goal)
    results = {}

    worker_map = {
        "market_research": market_research_worker,
        "finance": finance_worker,
        "operations": operations_worker,
        "legal": legal_worker,
        "hr": hr_worker
    }

    for task in tasks:
        if task in worker_map:
            results[task] = worker_map[task](goal)

    print("\n  WORKER REPORTS")
    for agent, data in results.items():
        print(f"\n  📌 {agent.replace('_', ' ').title()}")
        print(f"     → {data.get('summary', 'No summary')}")

    decision = make_final_decision(results)

    print("\n  FINAL DECISION")
    print(f"  Score: {decision['score']} | Verdict: {decision['verdict']}")

    return decision


# RUN
run_system("Evaluate feasibility of launching Product Y in Asia.")
```

Agent 1: Crisis Coordinator (Broadcaster)
- Broadcasts: “Data breach detected in customer database. Immediate response required.”
- Sends this to all other agents simultaneously.

🛡️ Agent 2: IT Security Agent
- Receives broadcast.
- Responds: “Isolate affected servers, patch vulnerabilities, start forensic analysis.”

📞 Agent 3: Communications Agent
- Receives broadcast.
- Responds: “Draft internal memo, prepare press release, notify stakeholders.”

💰 Agent 4: Finance Agent
- Receives broadcast.
- Responds: “Estimate financial impact, allocate emergency funds, review insurance coverage.”

👩‍⚖️ Agent 5: Legal Agent
- Receives broadcast.
- Responds: “Assess regulatory obligations, prepare compliance reports, advise on liability.”

👩‍💼 Agent 6: HR Agent
- Receives broadcast.
- Responds: “Brief employees, provide guidance on handling customer queries, ensure morale support.”

🧑‍⚖️ Agent 7: Decision Agent (Coordinator)
- Collects all responses.
- Integrates into a final crisis response plan:
“Servers isolated, communications prepared, financial impact assessed, compliance secured, employees briefed.”

```python

from openai import OpenAI
import json
import re


client = OpenAI(
    api_key="api key", 
    base_url="https://api.groq.com/openai/v1"
)

MODEL = "llama-3.3-70b-versatile"   


def safe_json_parse(text):
    try:
        return json.loads(text)
    except:
        match = re.search(r"\{.*\}", text, re.DOTALL)
        if match:
            try:
                return json.loads(match.group())
            except:
                pass
    return {"response": text}


def call_llm(system_prompt, user_message):
    response = client.chat.completions.create(
        model=MODEL,
        temperature=0,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_message}
        ]
    )
    return response.choices[0].message.content.strip()


def it_security_agent(message):
    reply = call_llm(
        system_prompt="""
        You are an IT Security Agent.
        Respond with JSON:
        {"action": "short response"}
        """,
        user_message=message
    )
    return safe_json_parse(reply)


def communications_agent(message):
    reply = call_llm(
        system_prompt="""
        You are a Communications Agent.
        Respond with JSON:
        {"action": "short response"}
        """,
        user_message=message
    )
    return safe_json_parse(reply)


def finance_agent(message):
    reply = call_llm(
        system_prompt="""
        You are a Finance Agent.
        Respond with JSON:
        {"action": "short response"}
        """,
        user_message=message
    )
    return safe_json_parse(reply)



def legal_agent(message):
    reply = call_llm(
        system_prompt="""
        You are a Legal Agent.
        Respond with JSON:
        {"action": "short response"}
        """,
        user_message=message
    )
    return safe_json_parse(reply)



def hr_agent(message):
    reply = call_llm(
        system_prompt="""
        You are an HR Agent.
        Respond with JSON:
        {"action": "short response"}
        """,
        user_message=message
    )
    return safe_json_parse(reply)


def crisis_coordinator():
    message = "Data breach detected in customer database. Immediate response required."

    print("\n🚨 Crisis Coordinator Broadcasting...")
    print(f"Message: {message}\n")

    return message


def decision_agent(responses):

    print("\n Decision Agent Processing...\n")

    combined = " ".join([r.get("action", "") for r in responses.values()])

    reply = call_llm(
        system_prompt="""
        You are a Decision Agent.
        Combine all responses into ONE clear crisis response plan.
        Return JSON:
        {"final_plan": "concise plan"}
        """,
        user_message=combined
    )

    return safe_json_parse(reply)


def run_crisis_system():

    # Step 1: Broadcast
    message = crisis_coordinator()

    # Step 2: Agents respond
    responses = {
        "IT Security": it_security_agent(message),
        "Communications": communications_agent(message),
        "Finance": finance_agent(message),
        "Legal": legal_agent(message),
        "HR": hr_agent(message)
    }

    print(" Agent Responses:")
    for agent, res in responses.items():
        print(f"\n🔹 {agent}: {res.get('action')}")

    # Step 3: Final Decision
    final = decision_agent(responses)

    print("\n FINAL CRISIS RESPONSE PLAN:")
    print(final.get("final_plan"))

    return final


run_crisis_system()
```

Scenario: Corporate Product Launch Broadcast
Imagine a company preparing to launch Product X in Q3. The Coordinator Agent (like a corporate program manager) sends out a broadcast announcement to all departments at once:
“Product X launch in Q3, target market North America, budget $5M.”


📢 Coordinator Agent (Broadcaster)
- Sends the launch announcement to all departments simultaneously.
- This is implemented in the code by the broadcast() function, which uses asyncio.gather() to run all agents in parallel.

📈 Marketing Agent
- Receives the broadcast.
- Responds with a marketing strategy: campaigns, channels, and positioning.
💰 Finance Agent
- Receives the broadcast.
- Responds with budget allocation and ROI forecasts.
🏭 Operations Agent
- Receives the broadcast.
- Responds with production and supply chain actions.
👩‍⚖️ Legal Agent
- Receives the broadcast.
- Responds with compliance checks and contract actions.
👩‍💼 HR Agent
- Receives the broadcast.
- Responds with staffing and training plans.

🧑‍⚖️ Decision Agent
- Collects all responses.
- Integrates them into a Final Corporate Launch Plan.
- In the code, this is the decision_agent() function that merges all outputs into one consolidated plan.

```python
# MULTI-AGENT CRISIS RESPONSE SYSTEM 

import asyncio
import json
import re
from openai import OpenAI

client = OpenAI(
    api_key="api key", 
    base_url="https://api.groq.com/openai/v1"
)

MODEL = "llama-3.3-70b-versatile"   


def safe_json_parse(text):
    try:
        return json.loads(text)
    except:
        match = re.search(r"\{.*\}", text, re.DOTALL)
        if match:
            try:
                return json.loads(match.group())
            except:
                pass
    return {"action": text.strip()}



async def call_llm(system_prompt, user_message):
    loop = asyncio.get_event_loop()
    response = await loop.run_in_executor(
        None,
        lambda: client.chat.completions.create(
            model=MODEL,
            temperature=0,
            messages=[
                {"role": "system", "content": system_prompt},
                {"role": "user", "content": user_message}
            ]
        )
    )
    return response.choices[0].message.content.strip()


async def it_security_agent(msg):
    reply = await call_llm("""
    You are an IT Security Agent.
    Respond in ONLY 1-2 words.
    Return JSON: {"action": "containment"}
    """, msg)
    return "IT Security", safe_json_parse(reply)


async def communications_agent(msg):
    reply = await call_llm("""
    You are a Communications Agent.
    Respond in ONLY 1-2 words.
    Return JSON: {"action": "alert"}
    """, msg)
    return "Communications", safe_json_parse(reply)


async def finance_agent(msg):
    reply = await call_llm("""
    You are a Finance Agent.
    Respond in ONLY 1-2 words.
    Return JSON: {"action": "impact analysis"}
    """, msg)
    return "Finance", safe_json_parse(reply)


async def legal_agent(msg):
    reply = await call_llm("""
    You are a Legal Agent.
    Respond in ONLY 1-2 words.
    Return JSON: {"action": "compliance check"}
    """, msg)
    return "Legal", safe_json_parse(reply)


async def hr_agent(msg):
    reply = await call_llm("""
    You are an HR Agent.
    Respond in ONLY 1-2 words.
    Return JSON: {"action": "employee briefing"}
    """, msg)
    return "HR", safe_json_parse(reply)


async def broadcast(message):
    print("\n🚨 Crisis Coordinator Broadcasting...")
    print(f"Message: {message}")

    tasks = [
        it_security_agent(message),
        communications_agent(message),
        finance_agent(message),
        legal_agent(message),
        hr_agent(message)
    ]

    results = await asyncio.gather(*tasks)
    return dict(results)


async def decision_agent(responses):

    print("\n Decision Agent Processing...\n")

    combined = " ".join([r.get("action", "") for r in responses.values()])

    reply = await call_llm("""
    You are a Decision Agent.

    Combine all actions into ONE short sentence (max 20 words).

    Return JSON:
    {"final_plan": "short crisis response plan"}
    """, combined)

    return safe_json_parse(reply)


async def run_system():

    message = "Data breach detected in customer database. Immediate response required."

    # Step 1: Broadcast
    responses = await broadcast(message)

    print("\n Agent Responses:\n")

    for agent, res in responses.items():
        print(f"🔹 {agent}: {res.get('action')}")

    # Step 2: Decision
    final = await decision_agent(responses)

    print("\n FINAL CRISIS RESPONSE PLAN:")
    print(final.get("final_plan"))

    return final


await run_system()

```
