```python
import pandas as pd

customers = pd.read_csv("Customers1 - Sheet1.csv")
sales = pd.read_csv("Sales1 - Sheet1.csv")
support = pd.read_csv("support1 - Sheet1.csv")

# Inspect datasets
print(customers.shape)
print(sales.shape)
print(support.shape)

print(customers.columns)
print(sales.columns)
print(support.columns)

print(customers.isnull().sum())
print(sales.isnull().sum())
print(support.isnull().sum())
```

```python
customers.columns = customers.columns.str.strip()
sales.columns = sales.columns.str.strip()
support.columns = support.columns.str.strip()
```

```python
customers["SignupDate"] = pd.to_datetime(customers["SignupDate"])
sales["OrderDate"] = pd.to_datetime(sales["OrderDate"])
```

```python
sales["Revenue"] = sales["Quantity"] * sales["Price"]
sales.head()
```

```python
import numpy as np

prices = sales["Price"].values
sales["Discounted_Price"] = prices * 0.90
```
